```python
r = requests.get("https://sprints.ai/")
```

```python
soup = BeautifulSoup(r.content, "html.parser")
soup
```

```python
title = soup.title.text.strip()
print("Title is :", title)
```

```python
testimonial_elements = soup.find_all("div",class_="position-relative")
```

```python
testimonial_elements
```

```python
testimonial_cards = soup.find_all('div', class_='testimonials-card')

for card in testimonial_cards:
    name = card.find('h4', class_='font-16').get_text(strip=True)
    title = card.find('span', class_='font-14').get_text(strip=True)
    rating_icons = card.find_all('i', class_='active')
    rating = len(rating_icons)
    comment = card.find('p', class_='mt-25').get_text(strip=True)
    
    testimonials.append({
        'name': name,
        'title': title,
        'rating': rating,
        'comment': comment
    })
```

```python
testimonials
```

```python
testimonials.remove('Mohamed El-Sayed')
```
